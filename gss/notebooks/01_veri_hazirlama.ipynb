```python

# 1. Dosya Yolları (VS Code dosya düzenine göre)
raw_path = '../data/raw/gss_2010_ham_full.parquet' # Parquet dosyanın bulunduğu yer
output_path = '../data/processed/temiz_veri.csv'

# 2. Belirlediğimiz 20 Kritik Sütun
hedef_sutunlar = [
    'year', 'realinc', 'income', 'educ', 'health', 'happy', 'class', 
    'prestg10', 'paeduc', 'age', 'sex', 'race', 'marital', 
    'polviews', 'trust', 'satfin', 'childs', 'wrkstat', 'region', 'natfare'
]

# 3. Veriyi Yükle ve Sütunları Seç
print("Veri yükleniyor...")
df = pd.read_parquet(raw_path)
df_sub = df[hedef_sutunlar].copy()

# 4. GSS Spesifik Temizlik (IAP, DK, NA Değerlerini NaN Yapma)
# GSS'de genellikle 0, 8, 9 ve bunların katları geçersiz veridir.
print("Geçersiz değerler temizleniyor...")
df_sub = df_sub.replace([0, 8, 9, 98, 99], np.nan)

# 5. Temiz Veriyi Kaydet
df_sub.to_csv(output_path, index=False)

print("-" * 30)
print(f"İşlem Tamamlandı!")
print(f"Kayıt Yeri: {output_path}")
print(f"Toplam Satır Sayısı: {len(df_sub)}")
print(f"Sütun Sayısı: {len(df_sub.columns)}")
```

```python
import numpy as np

# Baskılanacak sayısal sütunları seçiyoruz
numeric_cols = ['realinc', 'age', 'educ', 'prestg10', 'childs']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Alt sınırın altındakileri alt sınıra, üst sınırın üstündekileri üst sınıra eşitle
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

print("Aykırı değerler başarıyla baskılandı.")
```

#### Veri Hazırlığı ve Modelleme Stratejisi

```python
# natfare sütunundan target_money (1/0) hedef değişkenini üretme
df['target_money'] = np.where(df['natfare'] == 1, 1, 0)

# 1. Önce tüm hedef değişkenleri oluştur (Sıralama hatasını önlemek için)
df_sub['target_money'] = np.where(df_sub['natfare'] == 1, 1, 0)
df_sub['health_risk'] = df_sub['health'].apply(lambda x: 0 if x in [1, 2] else 1)

# 2. Veri sızıntısını ve hedef çakışmasını önlemek için ortak silme listesi
# Hem yeni hedefleri hem de bu hedeflerin türetildiği ham sütunları çıkarıyoruz
drop_cols = ['target_money', 'health_risk', 'happy', 'natfare', 'health']

# 1. Model: Yardım İhtiyacı
y_money = df_sub['target_money']
X_money = df_sub.drop(columns=drop_cols)

# 2. Model: Mutluluk Analizi
y_happy = df_sub['happy']
X_happy = df_sub.drop(columns=drop_cols)

# 3. Model: Sağlık Tahmini
y_health = df_sub['health_risk']
X_health = df_sub.drop(columns=drop_cols)

# Doğruluk kontrolü
print(df_sub[['natfare', 'target_money']].value_counts())
```

Anket sorularını $0$ ve $1$ şeklinde ikili (binary) sınıflara dönüştürdük:target_money: natfare "çok az" ise 1, değilse 0.health_risk: health "orta/kötü" ise 1, "iyi" ise 0.

```python
# 1. Hedef Değişkenleri Oluştur
df_sub['target_money'] = np.where(df_sub['natfare'] == 1, 1, 0)
df_sub['health_risk'] = df_sub['health'].apply(lambda x: 0 if x in [1, 2] else 1)

# 2. X ve y Ayrımı (Örnek: target_money tahmini için)
# Sızıntıyı önlemek için ham sütunları (natfare, health, happy) ve tüm hedefleri X'ten çıkarıyoruz.
targets_to_drop = ['target_money', 'health_risk', 'happy', 'natfare', 'health']
X = df_sub.drop(columns=targets_to_drop)
y = df_sub['target_money']

# 3. %80 Eğitim, %20 Test Ayrımı
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.20, 
    random_state=42, 
    stratify=y
)

# Kontrol çıktıları
print(f"Eğitim Seti (X_train) Boyutu: {X_train.shape}")
print(f"y_train Hedef Dağılımı:\n{y_train.value_counts(normalize=True)}")
```

Verinin %80'ini öğrenme (Eğitim), %20'sini ise başarı ölçümü (Test) için ayırdık. stratify=y kullanarak her iki sette de hedef oranının aynı kalmasını sağladık.

```python
# 1. Sayısal ve Kategorik sütunları ayır
sayisal_sutunlar = X_train.select_dtypes(include=['int64', 'float64']).columns
kategorik_sutunlar = X_train.select_dtypes(include=['object', 'category']).columns

# 2. Sayısal verileri Medyan ile doldur
for col in sayisal_sutunlar:
    medyan = X_train[col].median()
    X_train[col] = X_train[col].fillna(medyan)
    X_test[col] = X_test[col].fillna(medyan)

# 3. Kategorik verileri Mod ile doldur
for col in kategorik_sutunlar:
    mod = X_train[col].mode()[0]
    X_train[col] = X_train[col].fillna(mod)
    X_test[col] = X_test[col].fillna(mod)

# 4. Kontrol
print(f"X_train eksik veri: {X_train.isnull().sum().sum()}")
print(f"X_test eksik veri: {X_test.isnull().sum().sum()}")
```

Encoding

```python
# 1. Sadece 'object' veya 'category' tipindeki sütunları seç (sex bu listede olmayacak)
donusturulecekler = X_train.select_dtypes(include=['object', 'category']).columns

# 2. Seçilen sütunlara One-Hot Encoding uygula
X_train = pd.get_dummies(X_train, columns=donusturulecekler, drop_first=True)
X_test = pd.get_dummies(X_test, columns=donusturulecekler, drop_first=True)

# 3. Test setini eğitim seti sütunlarıyla eşitle (Eksik sütun kalmasın)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# 4. Son kontrol
print(f"Encoding sonrası toplam sütun sayısı: {X_train.shape[1]}")
print(f"Hala 'object' tipinde sütun var mı?: {(X_train.dtypes == 'object').any()}")
```

Bu kod, `select_dtypes` ile `sex` gibi zaten sayısal olan sütunları koruyup sadece metin tabanlı kategorik verileri belirleyerek sayısal forma dönüştürür. `One-Hot Encoding` işlemi sırasında `drop_first=True` kullanılarak gereksiz sütun karmaşası önlenirken, `reindex` adımıyla eğitim ve test setlerinin sütun yapıları birbirine eşitlenerek modelin teknik uyumsuzluk hatası vermesi engellenir.
