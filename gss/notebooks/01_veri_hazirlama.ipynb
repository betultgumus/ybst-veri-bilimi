```python

# 1. Dosya Yolları (VS Code dosya düzenine göre)
raw_path = '../data/raw/gss_2010_ham_full.parquet' # Parquet dosyanın bulunduğu yer
output_path = '../data/processed/temiz_veri.csv'

# 2. Belirlediğimiz 20 Kritik Sütun
hedef_sutunlar = [
    'year', 'realinc', 'income', 'educ', 'health', 'happy', 'class', 
    'prestg10', 'paeduc', 'age', 'sex', 'race', 'marital', 
    'polviews', 'trust', 'satfin', 'childs', 'wrkstat', 'region', 'natfare'
]

# 3. Veriyi Yükle ve Sütunları Seç
print("Veri yükleniyor...")
df = pd.read_parquet(raw_path)
df_sub = df[hedef_sutunlar].copy()

# 4. GSS Spesifik Temizlik (IAP, DK, NA Değerlerini NaN Yapma)
# GSS'de genellikle 0, 8, 9 ve bunların katları geçersiz veridir.
print("Geçersiz değerler temizleniyor...")
df_sub = df_sub.replace([0, 8, 9, 98, 99], np.nan)

# 5. Temiz Veriyi Kaydet
df_sub.to_csv(output_path, index=False)

print("-" * 30)
print(f"İşlem Tamamlandı!")
print(f"Kayıt Yeri: {output_path}")
print(f"Toplam Satır Sayısı: {len(df_sub)}")
print(f"Sütun Sayısı: {len(df_sub.columns)}")
```

```python
import numpy as np

# Baskılanacak sayısal sütunları seçiyoruz
numeric_cols = ['realinc', 'age', 'educ', 'prestg10', 'childs']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Alt sınırın altındakileri alt sınıra, üst sınırın üstündekileri üst sınıra eşitle
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

print("Aykırı değerler başarıyla baskılandı.")
```
